```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the customer transactions dataset
customer_transactions = pd.read_csv('cleaned_customer_transactions.csv')  
```

```python
# Brand color palette
brand_palette = ["#7A9EBA", "#4A6A8B", "#A5B5C9", "#E5E9F2", "#3C4F63"]
```

```python
# Visualization 2: Transaction Amount by Product Category with updated brand colors
plt.figure(figsize=(12, 6))
sns.boxplot(x='productcategory', y='amount', data=customer_transactions, hue='productcategory', palette=brand_palette, dodge=False)
plt.title("Transaction Amount by Product Category", color=brand_palette[4])
plt.xlabel("Product Category", color=brand_palette[3])
plt.ylabel("Transaction Amount", color=brand_palette[3])
plt.legend([],[], frameon=False)  # Remove the legend since hue is redundant here
plt.savefig("transaction_amount_by_category.png")  # Save the plot with brand styling
plt.show()
```
