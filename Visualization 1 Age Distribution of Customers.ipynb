```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the customer demographics dataset
customer_demographics = pd.read_csv('cleaned_customer_demographics.csv')
```

```python
# Create a palette based on the company website
brand_palette = ["#7A9EBA", "#4A6A8B", "#A5B5C9", "#E5E9F2", "#3C4F63", "#3976A4"]
```

```python
# Visualization 1: Age Distribution of Customers 
plt.figure(figsize=(10, 6))
sns.histplot(customer_demographics['Age'], bins=15, kde=True, color=brand_palette[5])  
plt.title("Age Distribution of Customers", color=brand_palette[4])
plt.xlabel("Age", color=brand_palette[3])
plt.ylabel("Frequency", color=brand_palette[3])
plt.savefig("age_distribution.png")  # Save the plot with brand styling
plt.show()
```
